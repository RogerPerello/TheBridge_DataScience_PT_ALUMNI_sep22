## Objetivos
1. Uso de bucle
2. Captura de datos por consola
3. Uso if-elif-else
4. Uso de try-except
5. Definición de funciones. Programación modular. 
6. Operadores lógicos.

```python
from random import choice


jugadas = ['piedra', 'papel', 'tijeras', 'lagarto', 'spock']
victorias_maquina = 0
victorias_jugador = 0
empates = 0
rondas_totales = 0


def rondas():
    print("¿Cuántas rondas quieres jugar?")
    while True:
        cantidad = input("Debes elegir un número impar: ")
        try:
            division = int(cantidad) % 2
        except ValueError:
            print("Tienes que introducir una cifra.")
            continue
        if division != 0:
            print(f"Número de rondas elegido: {cantidad}")
            return cantidad


def maquina():
    return choice(jugadas)


def jugador():
    while True:
        eleccion = input('¿Qué vas a sacar? ¿Piedra? ¿Papel? ¿Tijeras? ¿Lagarto? ¿O spock? ')
        if eleccion.lower() in jugadas:
            return eleccion
        else:
            print('Lo que has escrito no es correcto.')


def comparacion(maquina, jugador):
    if maquina == jugador:
        return 0
    elif maquina == 'piedra' and (jugador == 'papel' or jugador == 'spock'):
        return 2
    elif maquina == 'papel' and (jugador == 'tijeras' or jugador == 'lagarto'):
        return 2
    elif maquina == 'tijeras' and (jugador == 'piedra' or jugador == 'spock'):
        return 2
    elif maquina == 'lagarto' and (jugador == 'piedra' or jugador == 'tijeras'):
        return 2
    elif maquina == 'spock' and (jugador == 'papel' or jugador == 'lagarto'):
        return 2   
    else:
        return 1


def descripcion(jug_jugador, jug_maquina, result, rondas_tot, victorias_jug, victorias_maq, emp, num_rond):
    print(f"Has elegido {jug_jugador}.")
    print(f"Tu rival ha elegido {jug_maquina}.")
    if result == 0:
        print("Habéis empatado en esta ronda.")
    elif result == 1:
        print("Tu rival gana esta ronda.")
    else:
        print("Has ganado esta ronda.")
    print(f"Rondas jugadas: {rondas_tot} Rondas ganadas: {victorias_jug} Rondas perdidas: {victorias_maq} Empates: {emp}")
    if rondas_tot < num_rond:
        if rondas_tot == num_rond - 1:
            print(f"Queda {num_rond - rondas_tot} ronda.")
        else:
            print(f"Quedan {num_rond - rondas_tot} rondas.")
        print("-------------------------------------------------------------")


def intro():
    print("¡Bienvenido a la partida de piedra-papel-tijeras-lagarto-spock!")
    print("Para obtener la victoria, gana más rondas que tu rival.")


def final(vic_jugador, vic_maquina):
    print("¡Se acabó la partida!")
    if vic_jugador > vic_maquina:
        print(f"¡Felicidades! Has ganado {vic_jugador} a {vic_maquina}.")
    elif vic_maquina > vic_jugador:
        print(f"¡Lástima! Has perdido {vic_maquina} a {vic_jugador}.")
    else:
        print(f"¡Habéis empatado a {vic_jugador}! Podría estar mejor...")


intro()
num_rondas = int(rondas())
while rondas_totales != num_rondas:
    jugada_maquina = maquina()
    jugada_jugador = jugador()
    resultado = comparacion(jugada_maquina, jugada_jugador)
    if resultado == 0:
        empates += 1
    elif resultado == 1:
        victorias_maquina += 1
    else:
        victorias_jugador += 1
    rondas_totales += 1
    descripcion(jugada_jugador, jugada_maquina, resultado, rondas_totales, victorias_jugador, victorias_maquina, empates, num_rondas)
final(victorias_jugador, victorias_maquina)
```
