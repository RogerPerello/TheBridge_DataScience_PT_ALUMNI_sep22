### Ejercicio 2

### Ejercicio 5

Añadid una nueva columna al _dataframe_ con un valor booleano indicando una predicción básica de si en una cierta batalla el bando atacante será vencedor. Definiremos el valor como _True_ sí el bando atacante es mayor al defensor, y como _False_ en caso contrario. Definid como _NaN_ aquellos casos en los que no haya número de tropas en ningún bando. ¿En que casos, identificados por el número de batalla (*battle\_number*), nuestra predicción coincide con el resultado real?

```python
df_result = df_got.copy()

df_result['attacker_bigger'] = True

df_result.loc[((df_result['attacker_size'] <= df_result['defender_size'])) | (df_result['attacker_size'].isnull()), 'attacker_bigger'] = False

df_result.loc[(df_result['attacker_size'].isnull()) & (df_result['defender_size'].isnull()), 'attacker_bigger'] = None

df_result = df_result[(df_result['attacker_bigger'] == True) & (df_result['attacker_outcome'] == 'win')]

df_result.set_index('battle_number', inplace=True)

df_result = df_result[['attacker_size', 'defender_size', 'attacker_outcome', 'attacker_bigger']]

df_result
```
